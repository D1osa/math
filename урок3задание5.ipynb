```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
fig = plt.figure(figsize = (10,10))
ax = fig.add_subplot(1, 1, 1, projection = '3d')
x_val = np.arange(-5, 5, 0.5)
y_val = np.arange(-5, 5, 0.5)
x, y = np.meshgrid(x_val, y_val)
z = 2 * x**2 + 4 * y**2
ax.plot_wireframe(x, y, z)

x_1_val = np.arange(-5, 5, 0.5)
y_1_val = np.arange(-5, 5, 0.5)
x_1, y_1 = np.meshgrid(x_1_val, y_1_val)
z_1 = 2 * x_1**3 + 5 * y_1**2 + 140
ax.plot_wireframe(x_1, y_1, z_1, color = 'red')
```

```python
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
from matplotlib import cm
import matplotlib.pyplot as plt
fig = plt.figure(figsize = (10, 10))
ax = fig.add_subplot(1, 1, 1, projection = '3d')
xval = np.linspace(-10, 10, 100)
yval = np.linspace(-10, 10, 100)
x, y = np.meshgrid(xval, yval)
z = 5 * x**2 + y**3
surf = ax.plot_surface(x, y, z, rstride = 15, cstride = 10,cmap = cm.plasma)

xval1 = np.linspace(-10, 10, 100)
yval1 = np.linspace(-10, 10, 100)
x1, y1 = np.meshgrid(xval1, yval1)
z1 = 3 * x1**3 + y1**2 + 250
surf = ax.plot_surface(x1, y1, z1, rstride = 20, cstride = 20,cmap = cm.viridis)
```
