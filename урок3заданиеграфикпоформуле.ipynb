```python
import numpy as np
import matplotlib.pyplot as plt
y = lambda x: 1 * np.cos(2 * x + 3)
fig = plt.subplots()
x = np.linspace (-3, 3, 100)
plt.plot(x, y(x))
plt.show()
```
