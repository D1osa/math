```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import fsolve

x = np.linspace(-2, 3, 201)
plt.plot(x, np.exp(x)/x + 1 - x)
plt.plot(x, x**2 - 1)
plt.xlabel('x')
plt.ylabel('y')
plt.ylim(-1,5)
plt.grid(True)
plt.show()

def equations(p):
    x, y = p
    return (y - x**2 - 1, np.exp(x) + x * (1 - y) - 1)

x_1, y_1 = fsolve(equations, (-3, 3))
print(x_1, y_1)
```
