```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import math
```

```python
def convert(R, a):
    rad = a * np.pi/180
    x = R * np.cos(rad)
    y = R * np.sin(rad)
    return x, y

x, y = convert(2,45)
print(x, y)
```

```python
r = np.linspace(0, 2 * np.pi, 100)
x = np.cos(r)
y = np.sin(r)
plt.figure(figsize = (6, 6))
plt.plot(x, y)
plt.show()
```
